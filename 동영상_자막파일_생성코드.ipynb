```python
# 1. 필수 라이브러리 설치
!pip install openai-whisper googletrans==4.0.0-rc1

import whisper
from googletrans import Translator
import datetime
import os
from google.colab import files

def format_timestamp(seconds):
    td = datetime.timedelta(seconds=seconds)
    total_seconds = int(td.total_seconds())
    hours = total_seconds // 3600
    minutes = (total_seconds % 3600) // 60
    secs = total_seconds % 60
    millis = int(td.microseconds / 1000)
    return f"{hours:02}:{minutes:02}:{secs:02},{millis:03}"

def run_subtitle_generator():
    # 2. 동영상 파일 업로드 받기
    print("--- 단계 1: 동영상 파일을 업로드하세요 ---")
    uploaded = files.upload()
    if not uploaded:
        print("업로드된 파일이 없습니다.")
        return

    video_path = list(uploaded.keys())[0]
    output_filename = os.path.splitext(video_path)[0] + "_es.srt"

    # 3. Whisper 모델 로드 (GPU 사용)
    print(f"\n--- 단계 2: AI 모델 로딩 중 (Medium 모델) ---")
    model = whisper.load_model("medium")

    # 4. 음성 인식 (한국어)
    print(f"--- 단계 3: 음성 추출 및 텍스트 변환 시작 ---")
    result = model.transcribe(video_path, language="ko")

    # 5. 스페인어 번역 및 SRT 작성
    print(f"--- 단계 4: 스페인어 번역 중... ---")
    translator = Translator()

    with open(output_filename, "w", encoding="utf-8") as f:
        for i, segment in enumerate(result['segments']):
            start = format_timestamp(segment['start'])
            end = format_timestamp(segment['end'])
            text_ko = segment['text']

            try:
                # 한국어 -> 스페인어 번역
                text_es = translator.translate(text_ko, src='ko', dest='es').text
            except:
                text_es = text_ko

            f.write(f"{i + 1}\n{start} --> {end}\n{text_es}\n\n")

    print(f"\n--- 완료! 자막 파일이 생성되었습니다: {output_filename} ---")

    # 6. 결과 파일 자동 다운로드
    files.download(output_filename)

# 함수 실행
run_subtitle_generator()
```
